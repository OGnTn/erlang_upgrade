```python
import numpy
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
#Load every benchmark
mupi_benchmark = 'benchmarks_firefly/mupi'
#load all files in the mupi folder as csv and concatenate them, naming the columns after the files
mupi = pd.concat([pd.read_csv(f'{mupi_benchmark}/{file}', sep='\t', header=None, names=[file]) for file in os.listdir(mupi_benchmark)], axis=1)
mupi = mupi.reindex(sorted(mupi.columns, key=lambda x: int(x.split('_')[1].split('.')[0])), axis=1)
#extract the first row in to another dataframe
mupi_first = mupi.iloc[0]
#drop the first row from the original dataframe
mupi = mupi.drop(0)
#rename the columns to everything between the _ and the .
mupi.columns = [col.split('_')[1].split('.')[0] for col in mupi.columns]
#convert the columns to numeric

#plot every column in the dataframe as a boxplot with x axis label as the column name and y axis label as time
mupi.boxplot(figsize=(15,10))
plt.xlabel('Max Users Per Sub Instance')
plt.ylabel('Time (ms)')
plt.title('Max Users Per Sub Instance Benchmark')
plt.show()






```
